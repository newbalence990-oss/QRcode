```python
#安裝套件
import qrcode
from IPython.display import display

# 製程資訊
manufacturing_info = """
Product:YUNPID-2026-001
Batch:B24001
Machine:CNC-03
Date:2027-05-07
Operator:Yuhsin
"""

#生成二維條碼
gr = qrcode.make(manufacturing_info)

#儲存二維條碼
gr.save("manufacturing_qrcode.png")

#顯示二維條碼
display(gr)
```

```python
import barcode
from barcode.writer import ImageWriter
from IPython.display import Image, display

# 產品料號
product_id = input("小洪製程條碼展示-請輸入你設計的物料編號")

# [很重要]建立 Code128 條碼 （未來實務專题可以把code128改成廠商類型)
code = barcode.get('code128', product_id, writer=ImageWriter())

# 儲存條碼
filename = code.save("product_barcode")

# 顯示條碼
display(Image(filename=filename))
```
